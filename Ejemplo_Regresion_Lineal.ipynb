Definir un modelo -
Ajustar el modelo

```python
reg1 = linear_model.LinearRegression()
reg1.fit(x,y1)
```

Creamos otro modelo, pero lo ajustamos ahora a la edad y los ingresos al mismo tiempo

```python
reg2 = linear_model.LinearRegression()
reg2.fit(df.loc[:,["Edad","Ingresos"]].values,y2)
salida = reg2.predict([[30,63186]]) #Veamos que gasto deberia tener una persona que tiene 30 y recibe de ingresos 63186
salida
```

Separamos el conjunto de datos, cambiar el test_size segun porcentaje que deseas que tenga el conjunto de prueba

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(df, df['Gastos'],
                                                    test_size=0.05, random_state=42
                                                    )
```

```python
reg3 = linear_model.LinearRegression()
reg3.fit(X_train.loc[:,["Edad","Ingresos"]].values,y_train) #El ajuste se realiza solo con el 80% del conjunto de la fase anterior (train)
salidas = reg3.predict(X_test.loc[:,["Edad","Ingresos"]].values) #Probemos el modelo con el 20% del conjunto de la fase anterior (test)
```

```python
from sklearn.metrics import max_error
```
