```
ocurrencias = df.shape[0]
features = df.shape[1]
```

```
soporte = (1000/ocurrencias)
soporte
```

```
df = df.astype(str)
transactions = df.values.tolist()
transactions
```

```
# Training Apriori on the dataset
!pip install apyori
import apyori
from apyori import apriori
```

```
rules = apriori(transactions, min_support = soporte, min_confidence = 0.2, min_lift = 2, min_length = 2)
```

```
results = list(rules)
```

```
results
```

```
def inspect(results):
    rh          = [tuple(result[2][0][0]) for result in results]
    lh          = [tuple(result[2][0][1]) for result in results]
    supports    = [result[1] for result in results]
    confidences = [result[2][0][2] for result in results]
    lifts       = [result[2][0][3] for result in results]
    return list(zip(rh, lh, supports, confidences, lifts))
```

```
resultDataFrame=pd.DataFrame(inspect(results),
                columns=['rhs','lhs','support','confidence','lift'])
```
