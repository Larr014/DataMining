```python
import pandas as pd
```

```python
df = pd.read_csv("/content/transacciones.csv")
df
```

```python
df.describe()
```

```python
productos = df['Producto'].unique()
productos
```

```python
transaccion = df['Transaccion'].unique()
transaccion
```

```python
d = []
for x in transaccion:
  #d1 = [str(x)]
  d1 = []
  t = df[df['Transaccion']==x]['Producto']
  d1.extend(list(t))
  d.append(d1)
d #2
```

```python
type(d)
```

```python
!pip install apyori
import apyori
from apyori import apriori as apriori2
```

```python
reglas = apriori2(transactions=d, min_support = 0.2, min_confidence = 0.2,  max_length=5)
```

```python
resultados = list(reglas)
resultados
```

```python
def inspect(results):
    rh          = [tuple(result[2][0][0]) for result in results]
    lh          = [tuple(result[2][0][1]) for result in results]
    supports    = [result[1] for result in results]
    confidences = [result[2][0][2] for result in results]
    lifts       = [result[2][0][3] for result in results]
    return list(zip(rh, lh, supports, confidences, lifts))
```

```python
resultadosFinales=pd.DataFrame(inspect(resultados),
                columns=['rhs','lhs','support','confidence','lift'])
```
