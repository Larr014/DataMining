```python
import pandas as pd
```

```python
df = pd.read_csv('/content/HW2.csv')
df
```

```python
df =df.drop(["title","culture","dateOfBirth","DateoFdeath","mother","father","heir","spouse","house","book1","book2","book3","book4","book5","isAliveMother","isAliveFather","isAliveHeir","isAliveSpouse","age","isPopular","boolDeadRelations","S.No","plod","name"],axis=1)
df
```

```python
df = df.dropna()
df
```

```python
X = df.copy()
X = X.drop(['isAlive'], axis=1)
X
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, df['isAlive'],
                                                    test_size=0.2, random_state=6#42
                                                    )
```

Se implementa el algoritmo

Random Forest

```python
type(X_test)
```

```python
from sklearn.ensemble import RandomForestClassifier
clf = RandomForestClassifier(random_state=0)
clf.fit(X_train, y_train)
y_pred = clf.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, multilabel_confusion_matrix, ConfusionMatrixDisplay, confusion_matrix
confusion = multilabel_confusion_matrix(y_test,y_pred)
print(f"acc: {accuracy_score(y_test,y_pred)}")
print(f"precision: {precision_score(y_test,y_pred,average='macro')}")
print(f"recall: {recall_score(y_test,y_pred,average='macro')}")
print(f"f1: {f1_score(y_test,y_pred,average='micro')}")
confusion_display = ConfusionMatrixDisplay(confusion[0],display_labels=['Sobrevive','No Sobrevive']).plot(values_format='d')
```

DT

```python
from sklearn import tree
from subprocess import check_call
from IPython.display import Image as PImage
decision_tree = tree.DecisionTreeClassifier(criterion='entropy',
                                            min_samples_split=20,
                                            min_samples_leaf=3,
                                            max_depth = 4,
                                            class_weight={1:3.5})
decision_tree.fit(X_train, y_train)

# exportar el modelo a archivo .dot
with open(r"tree1.dot", 'w') as f:
     f = tree.export_graphviz(decision_tree,
                              out_file=f,
                              max_depth = 7,
                              impurity = True,
                              feature_names = list(df.drop(['isAlive'], axis=1)),
                              class_names = ['No Sobrevive', 'Sobrevive'],
                              rounded = True,
                              filled= True )

# Convertir el archivo .dot a png para poder visualizarlo
check_call(['dot','-Tpng',r'tree1.dot','-o',r'tree1.png'])
PImage("tree1.png")
```

```python
y_pred2 = decision_tree.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, multilabel_confusion_matrix, ConfusionMatrixDisplay, confusion_matrix
confusion = multilabel_confusion_matrix(y_test,y_pred2)
print(f"acc: {accuracy_score(y_test,y_pred2)}")
print(f"precision: {precision_score(y_test,y_pred2,average='macro')}")
print(f"recall: {recall_score(y_test,y_pred2,average='macro')}")
print(f"f1: {f1_score(y_test,y_pred2,average='micro')}")
confusion_display = ConfusionMatrixDisplay(confusion[0],display_labels=['Sobrevive','No Sobrevive']).plot(values_format='d')
```
